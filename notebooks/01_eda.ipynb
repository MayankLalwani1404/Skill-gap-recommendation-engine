# CareerMap AI: Exploratory Data Analysis

This notebook explores the O*NET datasets to understand the distribution of skills, job zones, and basic career clusters.

```python
import os
import sys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sys.path.insert(0, os.path.abspath("../"))
from src.data_preprocessing import get_processed_data

role_df, long_df, role_skills = get_processed_data()
print(f"Total Occupations: {len(role_df)}")
print(f"Total Unique Skills: {long_df['skill_clean'].nunique()}")
```

## 1. Top Skills Across All Occupations

```python
top_skills = long_df.groupby("skill_clean")["onet_code"].nunique().sort_values(ascending=False).head(20)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_skills.values, y=[s.title() for s in top_skills.index], palette="viridis")
plt.title("Top 20 Most Demanded Skills in O*NET")
plt.xlabel("Number of Occupations Requiring Skill")
plt.show()
```

## 2. Job Zone Distribution

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=role_df, x="job_zone", palette="coolwarm")
plt.title("Occupations by Job Zone (Experience Level)")
plt.xlabel("Job Zone (1=Entry, 5=Extensive Prep)")
plt.ylabel("Count")
plt.show()
```
